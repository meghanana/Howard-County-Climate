```python
import pandas as pd
import matplotlib.pyplot as plt

dftemp = pd.read_csv("/Users/megh/code/project/data.csv", comment='#')

dftemp['Temperature'] = pd.to_numeric(dftemp['Temperature'], errors='coerce')

# Plotting
plt.figure(figsize=(16, 6))
plt.plot(dftemp['Year'], dftemp['Temperature'], linestyle='--', marker='o', color='red')

plt.title('Howard County: Average Temp Change in Jan-July (20002025) F', fontsize=20)
plt.xlabel('Year')
plt.xticks(dftemp['Year'], rotation=90)
plt.ylabel('Temperature (F)')
plt.grid(True, linestyle='--', alpha=1)
plt.tight_layout()
plt.show()

def mean(dftemp):
    return sum(dftemp.Temperature) / len(dftemp.Temperature)
print("Mean Value:", mean(dftemp))

x = dftemp['Year']
y = dftemp['Temperature']

slope, intercept = np.polyfit(x, y, 1) 
line = slope * x + intercept  

plt.figure(figsize=(16,6))
plt.plot(x, y, 'o--', color='red', label='Actual Data')   # your data
plt.plot(x, line, color='green', linewidth=2, label='Line of Best Fit')


if slope > 0:
    print("The Temperature is increasing over time.")
    print('This indicates an increase of the overall climate temperature in Howard County the past 25 years')
elif slope < 0:
    print("The Temperature is decreasing over time.")
    print('This indicates an decrease of overall climate temperature in Howard Count the past 25 years.')

else:
    print("The Temperature is stable over time.")
    print('This indicates an stable climate temperature in Howard County the past 25 years.')





```

```python
import pandas as pd
import matplotlib.pyplot as plt

dfperc = pd.read_csv("/Users/megh/code/project/percipitation.csv", comment='#')

print("Columns in the dataframe:", dfperc.columns)

dfperc['Value'] = pd.to_numeric(dfperc['Value'], errors='coerce')


plt.figure(figsize=(16, 6))
plt.plot(dfperc['Date'], dfperc['Value'], linestyle='--', marker='o', color='Blue')

plt.title('Howard County: Average Percipitation Change in Jan-July (20002025) in F', fontsize=20)
plt.xlabel('Date')
plt.xticks(dfperc['Date'], rotation=90)
plt.ylabel('Precipitation (inches)')
plt.grid(True, linestyle='--', alpha=1)
plt.tight_layout()
plt.show()

def mean(dfperc):
    return sum(dfperc.Value) / len(dfperc.Value)
print("Mean Value:", mean(dfperc))

x = dfperc['Date']
y = dfperc['Value']

slopeperc, intercept = np.polyfit(x, y, 1) 
line = slopeperc * x + intercept  

plt.figure(figsize=(16,6))
plt.plot(x, y, 'o--', color='blue', label='Actual Data')   # your data
plt.plot(x, line, color='green', linewidth=2, label='Line of Best Fit')

if slopeperc > 0:
    print("The precipitation is increasing over time.")
elif slopeperc < 0:
    print("The precipitation is decreasing over time.")
else:
    print("The precipitation is stable over time.")
```
